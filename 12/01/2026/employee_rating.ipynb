```python
import numpy as np
import pandas as pd
import random 

ratings = np.random.randint(1,6, size=(100, 4))
print("Original ratings\n")
print(ratings)

min_val = ratings.min()
max_val = ratings.max()
print("\nMinimum value in the ratings:")
print(min_val)
print("\nMaximum value in the ratings:")
print(max_val)


normalized_ratings = (ratings - min_val)/(max_val - min_val)
print("\nNormalize ratings:")
print(normalized_ratings)


average_per_employee = ratings.mean(axis=1)
print("\nAverage rating value:")
print(average_per_employee)

company_rating = average_per_employee.mean()
print("\naverage company rating:")
print(company_rating)

company_employee_avg = np.where(average_per_employee > company_rating)
print("\nAverage rating greater than employee:")
print(company_employee_avg[0])
```

```python
data = {
    "empid" : [1, 2, 3, 4, 5, 6],
    "department" : ["HR", "IT","Finance", "IT", "HR", "IT"],
    "quator": [1, 3, 1, 1, 4, 2],
    "performance_score" : [5, 4, 5, 4, 4, 3]

}

df = pd.DataFrame(data)
print(df)
```

```python
group_dept = df.groupby("department")
print("\nGroup by department:")
print(group_dept.mean())

group_avg_quator = df.groupby(["department","quator"])["performance_score"].mean()
print("\n Quarter wise average: ")
print("\n",group_avg_quator)


pivot_df = df.pivot_table(
    index="department",
    columns="quator",
    values="performance_score",
    aggfunc="mean"
)

print("\nPivot Table:")
print(pivot_df)
```
