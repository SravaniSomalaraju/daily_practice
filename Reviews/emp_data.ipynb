```python
from faker import Faker
import pandas as pd
import random

fake = Faker()

data=[]

for i in range(1, 11):
    data.append({
        "id" : i,
        "name" : fake.name(),
        "salary" : random.randint(20000, 80000)
    })


df = pd.DataFrame(data)

print(df)
```

```python
df["bonus"] = df["salary"] * 0.10

print("employee salary after adding bonus: ")
print(df)
```

```python
high_salary = df[df["salary"] > 50000]

print("salary greater than 50000 are : ")
print(high_salary)
```
