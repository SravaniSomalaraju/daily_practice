```python
job_count = df['Job Title'].value_counts()
job_count
```

```python
import matplotlib.pyplot as plt

top_jobs = df['Job Title'].value_counts().head(10)

plt.figure()
plt.bar(top_jobs.index, top_jobs.values)
plt.xlabel("Job Title")
plt.ylabel("Employee Count")
plt.title("Top 10 Job Titles")
plt.show()


```

```python
gender_counts = df['Sex'].value_counts()
gender_counts

plt.figure()
plt.pie(gender_counts, labels=['female', 'male'], autopct='%1.1f%%')
plt.title("Gender of Employees")
plt.legend()
plt.show()
```
